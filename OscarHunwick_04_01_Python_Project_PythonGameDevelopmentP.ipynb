```python
#CREATING THE OBJECTS/INTERFACE AND HOW IT LOOKS
# Create two paddles, A on the left and B on the right of the screen


#Create a ball


#Create an event to move the paddle vertically on pressing a certain key


#Create the function to update the score after each player misses a collision 



#Import the turtle library 

import turtle 


from turtle import *
import turtle  

#Create the screen/interface

sc = turtle.Screen()                                              #Assign name to variable 
sc.setup(width=1050, height=650)                                                #Setting up the size of the window 
turtle.bgcolor("black")                                           #Allows you to set the background colour 
turtle.title("Turtle Pong")                                       #Sets the game title at the top of the window 

#Input the word PONG onto the background
sketch = turtle.Turtle()                                          #Assign a name to the variable 
sketch.speed(0)                                                   #Sets the speed of the drawing 
sketch.color("white")                                             #Sets the colour of the drawing 
sketch.penup()                                                    #' picks up' the pen so the turtle will no longer draw
sketch.hideturtle()                                               #Hides the turtle so it does not show on the screen after completing the drawing
sketch.goto(5,240)                                               #tells the turtle where to go to do the drawing
sketch.write("P O N G",                                           #tells the turtle what to write 
             align = "center", font=( "Impact", 50, "bold"))     #formats the text assignment, chooses the font, font size, and makes it bold

                                                                 #the same structure is used to create the paddles and ball
#Create left paddle

left_pad = turtle.Turtle()                                        
left_pad.speed(0)                                                 
left_pad.color("red")                                             
left_pad.shape("square")                                          
left_pad.penup()                                                  
left_pad.goto(-475,0)                                             
left_pad.shapesize(stretch_wid=6, stretch_len=1)                  

#Create the right paddle object

right_pad = turtle.Turtle()                                       
right_pad.speed(0)                                                
right_pad.color("blue")                                           
right_pad.shape("square")                                         
right_pad.penup()                                                 
right_pad.goto(475,0)                                             
right_pad.shapesize(stretch_wid=6, stretch_len=1)                 

#Create the ball object 

ball = turtle.Turtle()                                            
ball.speed(2)                                                      
ball.shape("circle")                                              
ball.color("light green")
ball.shapesize(stretch_wid=1.5, stretch_len=1.5)
ball.dx = 5
ball.dy = -5
ball.right(90)
ball.forward(300)
ball.right(180)
ball.forward(600)
ball.right(180)
ball.forward(300)
ball.penup()


#Scoring

left_player = 0                                                                 #input: defines the score for each player
right_player = 0                                                                #output: score at the top set to zero for start


#Display the score

sketch = turtle.Turtle()                                                       #same structure used as the paddles and ball 
sketch.speed(0)
sketch.color("white")
sketch.penup()
sketch.hideturtle()
sketch.goto(0, 270)
sketch.write("RED 0                                   BLUE 0",
            align= "center", font=("Courier", 24, "normal"))



#FUNCTIONS FOR MOVEMENT
                                                                            #input
def paddleaup():                                                            # 4 functions defined as paddleaup and paddleadown and same for the second paddle
    y = left_pad.ycor()                                                     #gets the current y-coordinare of the left paddle
    y += 20                                                                 #decreases or increases the y-coordinate by 20 units to move the paddle
    left_pad.sety(y)                                                        #sets the new y-coordinate for the left paddle
                                                                            #output: using the same code but with different values allowing the paddle to move up and down
def paddleadown():
    y = left_pad.ycor()
    y -= 20
    left_pad.sety(y)

def paddlebup():
    y = right_pad.ycor()
    y += 20
    right_pad.sety(y)
 
 
def paddlebdown():
    y = right_pad.ycor()
    y -= 20
    right_pad.sety(y)

# Keyboard bindings                                                        #input
sc.listen()                                                                #enables listening for when a key is pressed
sc.onkeypress(paddleaup, "w")                                              #binds the function paddleaup to the "w" key for moving the left paddle upward
sc.onkeypress(paddleadown, "s")                                            #the same is used to then be able to bind a key to up and down for the paddles
sc.onkeypress(paddlebup, "Up")                                             #output: when the assigned key is pressed, the paddle moves in the designated direction
sc.onkeypress(paddlebdown, "Down") 

while True:
    sc.update()                                                            #updates the screen to reflect change(refresh rate)
    ball.setx(ball.xcor()+ball.dx)                                         #moves the ball horizontally by adding the value of dx to its current x-coordinate
    ball.sety(ball.ycor()+ball.dy)                                         #moves the ball vertically by adding dy to its current y-coordinate
                                                                           #dx and dy reflect the amount of change in an object's position
    #making borders so the ball doesn't disappear into the abyss of turtle
                                                                           #input:
    if ball.ycor() > 300:                                                  #checks if the ball has reached the top border
        ball.sety(300)                                                     #if it has reached the top, sets the y-coord of the ball to the top border
        ball.dy *= -1                                                      #output: reverses the vertical direction of the ball by multiplying the dy by -1, causing it to start moving downwards
    
    if ball.ycor() < -300:                                                 #the same for the bottom border
        ball.sety(-300)
        ball.dy *= -1
                                                                                          #input:
    if ball.xcor() > 550:                                                                 #checks if the ball has gone past the right border
        ball.goto(0, 0)                                                                   #resets the ball to center if it has passed it 
        ball.dx *= -1                                                                     #reverses the horizontal direction of the ball so the game has alternating 'kick off's'
        left_player += 1                                                                  #output: increases the left player score by 1 
        sketch.clear()                                                                    #output: clears the previous score(e.g. 0-0 is wiped and re-written as 1-0)
        sketch.write("RED : {}                                     BLUE: {}".format(      #output: updates the score display for both players
                        left_player, right_player), align="center",
                        font=("Courier", 24, "normal"))
 
    if ball.xcor() < -550:                                                                #same as above for if the left player scores
        ball.goto(0, 0)
        ball.dx *= -1
        right_player += 1
        sketch.clear()
        sketch.write("RED : {}                                     BLUE: {}".format(
                        left_player, right_player), align="center",
                        font=("Courier", 24, "normal"))
 
    # Paddle/ Ball collision                                                              #input: checks if the ball is colliding with the paddle
    if (ball.xcor() > 455 and ball.xcor() < 475) and \
        (ball.ycor() < right_pad.ycor() + 60 and ball.ycor() > right_pad.ycor() - 60):    
        ball.setx(455)                                                                    #if it has, the x-coord is set to the position where it collided
        ball.dx *= -1                                                                     #output: reverses the ball direction 
                                                                                          #repeated for the right paddle
    if (ball.xcor() < -455 and ball.xcor() > -475) and \
        (ball.ycor() < left_pad.ycor() + 60 and ball.ycor() > left_pad.ycor() - 60):
        ball.setx(-455)
        ball.dx *= -1



   

turtle.done()
turtle.showturtle()

help(turtle)
```
